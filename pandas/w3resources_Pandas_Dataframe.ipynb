49. Write a Pandas program to append data to an empty DataFrame.


```python
import pandas as pd

# Create an empty DataFrame
empty_df = pd.DataFrame(columns=['Name', 'Age'])

# Data to append
data_to_append = {'Name': ['Alice', 'Bob', 'Charlie'],
                  'Age': [25, 30, 35]}

# Create a DataFrame from the data to append
data_df = pd.DataFrame(data_to_append)

# Append the data to the empty DataFrame
empty_df = empty_df.append(data_df, ignore_index=True)

# Display the resulting DataFrame
print(empty_df)
```
